exercise 2
x*log(x)

```python
from scipy.misc import derivative
```

```python
def xlog(x):
    return x*np.log(x)

x = np.linspace(1, 100, 1000)
plt.plot(x, xlog(x))
```

now the derivative

```python
log_derivative = derivative(xlog, x, dx=1e-6)
plt.plot(x, log_derivative)
```

exercise 3
e**x*log(x)+5

```python
def elog(x):
    return (np.e**x*np.log(x) + 5)

x = np.linspace(1, 100, 1000)
plt.plot(x, elog(x))
```

now the derevitavi

```python
e_der = derivative(elog, x, dx=1e-6)

plt.plot(x, e_der)
```

exercise 4
sin(1/2*x) + 2
interval [0, 2pi]

```python
y =np.sin(1/2*x)+2
x = np.linspace(0, 2*np.pi, 1000)
```

```python
def sin(x):
    return np.sin(1/2*x) + 2

sin_der = derivative(sin, x, dx=1e-6)
plt.plot(x, y)
plt.plot(x, sin_der)
plt,plt.legend(["sin", "derivative of sin"])
```
